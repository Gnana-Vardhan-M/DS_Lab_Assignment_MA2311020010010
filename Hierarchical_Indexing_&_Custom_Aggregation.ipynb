```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# Load regional_sales.csv with columns
data = {
    "Region": np.random.choice(["North", "South", "East", "West"], 100),
    "ProductCategory": np.random.choice(["Electronics", "Clothing", "Grocery"], 100),
    "Month": np.random.choice(["Jan", "Feb", "Mar"], 100),
    "Sales": np.random.randint(1000, 10000, 100),
    "Profit": np.random.randint(100, 2000, 100)
}
df = pd.DataFrame(data)
df.to_csv("regional_sales.csv", index=False)

df = pd.read_csv("regional_sales.csv")

files.download("regional_sales.csv")
```

```python
# Set a hierarchical index using Region and ProductCategory
df.set_index(["Region", "ProductCategory"], inplace=True)
```

```python
# Calculate total Sales and Profit for each Region-ProductCategory pair
summary = df.groupby(level=[0, 1])[["Sales", "Profit"]].sum()
```

```python
# Create a new column ProfitMargin = (Profit / Sales) × 100
summary["ProfitMargin"] = (summary["Profit"] / summary["Sales"]) * 100
```

```python
# Find the Region-ProductCategory combination with the highest ProfitMargin
max_idx = summary["ProfitMargin"].idxmax()
max_val = summary["ProfitMargin"].max()
print("\nHighest Profit Margin Combination:")
print("Region-ProductCategory:", max_idx)
print("ProfitMargin:", max_val)
```

```python
print("Highest ProfitMargin at:", max_idx)
print("ProfitMargin:", max_val)
print(summary.reset_index())
```
