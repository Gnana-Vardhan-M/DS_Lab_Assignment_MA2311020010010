```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# You have two files: student_personal.csv and exam_scores.csv
students = pd.DataFrame({
    "StudentID": range(1, 21),
    "Name": [f"Student{i}" for i in range(1, 21)],
    "Department": np.random.choice(["CSE", "ECE", "MECH"], 20),
    "Year": np.random.choice([1, 2, 3, 4], 20)
})
students.to_csv("student_personal.csv", index=False)

scores = pd.DataFrame({
    "StudentID": np.random.randint(1, 21, 50),
    "Subject": np.random.choice(["Math", "Physics", "CS"], 50),
    "Marks": np.random.randint(40, 100, 50),
    "ExamType": np.random.choice(["Midterm", "Final"], 50)
})
scores.to_csv("exam_scores.csv", index=False)

students = pd.read_csv("student_personal.csv")
scores = pd.read_csv("exam_scores.csv")

files.download("student_personal.csv")
files.download("exam_scores.csv")
```

```python
# Merge both files on StudentID
df = pd.merge(students, scores, on="StudentID")
```

```python
# Calculate the average marks per Department for each Subject
avg_marks = df.groupby(["Department", "Subject"])["Marks"].mean().reset_index()
```

```python
# Reshape the result so that each Subject becomes a column
pivot_table = avg_marks.pivot(index="Department", columns="Subject", values="Marks")
df.head()
```

```python
# Display the final pivoted table with Department as index
print(pivot_table)
```
